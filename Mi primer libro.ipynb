#### El coeficiente de determinacin ajustado:

$$R^2_{adj} = 1 - \frac{1 - (R^2\cdot(n - 1))}{n - m - 1}$$
                    
donde $n$ es el tamalo de la muestra, $m$ es el nuemro d evariables independientes y$R^2$ es le coeficiente de determinacin

```python
# Definimos R2_adj
#X_test.shape[1] es para saber el numero de columnas

r2_adj = 1 - (r2 * (len(y_test) - 1)) / (len(y_test) - X_test.shape[1] - 1)
r2_adj
```
